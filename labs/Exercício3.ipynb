```python
import pandas as pd
import numpy as np
```

```python
premierLeague2016_2017Raw = pd.read_csv('http://www.football-data.co.uk/mmz4281/1617/E0.csv')
premierLeague2016_2017Raw.shape
```

```python
premierLeague2016_2017Raw.head()
```

```python
pl = premierLeague2016_2017Raw.loc[ : , 'Date':'FTAG']
pl.Date = pd.to_datetime(pl.Date)
pl.head()
```

```python
pl.insert(0,'Match', (range(1, len(pl) + 1)))
pl.head()
```

```python
pl['Location'] = pl['HomeTeam']
```

```python
id_vars=['Match', 'Date', 'Location', 'HomeTeam', 'AwayTeam']
value_vars=['FTHG', 'FTAG']

plwide = pd.melt(pl, id_vars=id_vars, value_vars=value_vars, var_name='GoalPlace', value_name='GoalNum')
teamSeries = plwide.apply(lambda r: r['HomeTeam'] if r['GoalPlace'] == 'FTHG' else r['AwayTeam'], axis=1)
plwide.insert(3, 'Team', teamSeries)
plwide = plwide.sort_values(['Match', 'GoalPlace'], ascending=[True, False]).reset_index()
plwide = plwide.drop(['HomeTeam', 'AwayTeam', 'GoalPlace', 'index'], axis=1)
plwide.head()
```
